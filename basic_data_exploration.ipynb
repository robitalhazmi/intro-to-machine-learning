## Using Pandas to Get Familiar With Your Data

```python
import pandas as pd
```

Baris kode di atas mengimpor pustaka Pandas dan memberikan alias “pd” untuk memudahkan penggunaan selanjutnya.

```python
# save filepath to variable for easier access
melbourne_file_path = './melb_data.csv'
# read the data and store data in DataFrame titled melbourne_data
melbourne_data = pd.read_csv(melbourne_file_path) 
# print a summary of the data in Melbourne data
melbourne_data.describe()
```

1. Menyimpan Lokasi File: Kode pertama menyimpan jalur file CSV ke dalam variabel bernama melbourne_file_path. Jalur file ini relatif terhadap direktori kerja saat ini dan mengacu pada file “melb_data.csv”.
2. Membaca Data: Kode kedua menggunakan pustaka pandas untuk membaca data dari file CSV. Data tersebut disimpan dalam sebuah DataFrame yang diberi nama melbourne_data.
3. Ringkasan Data: Kode ketiga mencetak ringkasan data dari DataFrame melbourne_data menggunakan metode describe(). Ringkasan ini berisi statistik (seperti jumlah data, rata-rata, nilai minimum, maksimum, dll.) untuk setiap kolom numerik dalam DataFrame. Informasi ini berguna untuk memahami distribusi dan statistik dasar dari data tersebut.
